#### How to plot Contour and Surface Plots ####

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# Create a dataset of X and Y values

X = np.array([2,1,4,5,3,6,2,5]).reshape((-1,1))
Y = np.array([1,2,3,5,3,6,4,2]).reshape((-1,1))
```

```python
# The applicable function for this example

func =  X**2 + Y**2
```

```python
# Create a meshgrid - this is a function in numpy

x = np.arange(0,6,0.5)
y = np.arange(0,6,0.5)
Xl, Yl = np.meshgrid(x, y)
Xl.shape,Yl.shape
```

```python
# Plot the grid you just created

plt.plot(Xl,Yl, marker='.', color='k', linestyle='none')
plt.show()
```

```python
# Prepare the Z axis value based on the formula X^2 + Y^2
# Reshape it to the shape of the grid

Z =  np.array(Xl**2+Yl**2).reshape(Xl.shape)
Z.shape

# df = pd.DataFrame(np.hstack((Xl.flatten()[:,np.newaxis],Yl.flatten()[:,np.newaxis])))
# df.columns = ['x1','y1']
# Z =  np.array(df['x1']**2 + df['y1']**2).reshape(Xl.shape)
```

```python
# Create a surface plot and projected filled contour plot under it.

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(Xl, Yl, Z, rstride=3, cstride=3, linewidth=1, antialiased=True,
                cmap='viridis')

cset = ax.contourf(Xl, Yl, Z, zdir='z', offset=-0.15, cmap='viridis')

# rotate the axes and update
angle = 120
a = ax.view_init(15,angle)
```

```python
# The contour plot - plot the data points first and then
# plot the contour based on xl,yl,Z

plt.scatter(X,Y)
CS = plt.contour(Xl,Yl,Z,10)
plt.clabel(CS, inline=1, fontsize=10)
plt.xlim(0,6)
plt.ylim(0,6)
plt.xlabel("x1")
plt.ylabel("y1")
```
