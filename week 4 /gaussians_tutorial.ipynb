```python
import numpy as np
from numpy.random import randn
from numpy.random import rand
from numpy.random import seed
import matplotlib.pyplot as plt
from scipy.stats import boxcox
seed(1)
```

```python
mu=1.0
sigma = 1
data1 = np.random.normal(mu,sigma,1000)
plt.hist(data1,100,density=True)
count,bin = np.histogram(data1,100,)

#print(count,bin)

plt.plot(bin, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bin - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r')
plt.scatter(2.3,0,c='r')
plt.show()
```

```python
# Plotting Gaussians

plt.xlabel("X")
plt.ylabel("Density")
plt.plot(bin, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bin - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r')
plt.plot(bin, 1/(sigma/2 * np.sqrt(2 * np.pi)) * np.exp( - (bin - mu)**2 / (2 * sigma/2**2) ),linewidth=2, color='g')
plt.plot(bin, 1/(sigma*5 * np.sqrt(2 * np.pi)) * np.exp( - (bin - mu)**2 / (5 * 2*sigma**2) ),linewidth=2, color='b')
plt.plot(bin, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bin - 3*mu)**2 / (2*sigma**2) ),linewidth=2, color='y')
plt.annotate('sigma:0.5,mu:1',xy=(1.3,0.7))
plt.annotate('sigma:1',xy=(-1,0.2))
plt.annotate('sigma:5',xy=(0.1,0.1))
plt.annotate('sigma:1,mu:3',xy=(3,0.42))
plt.show()
```

```python
# # generate a univariate data sample
data2 = 5 * randn(100) + 10
tail = 10 + (rand(50) * 100)

# # add long tail
data2 = np.append((data2**2), tail)
count1,bin1,ignore = plt.hist(boxcox(data2),5)
plt.show()
```

```python
ad =2,-2
mean = [0, 0]
cov = [[1, 0], [0, 1]]  # diagonal covariance
x, y = np.random.multivariate_normal(mean, cov,2000).T
plt.plot(x, y, 'x')
plt.scatter(ad[0],ad[1],c='r')
plt.axis('equal')
plt.show()
```
